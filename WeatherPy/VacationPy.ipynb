## Starter Code to Import Libraries and Load the Weather and Coordinates Data

```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
import numpy as np
import holoviews as hv
hv.extension('bokeh')

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
city_data_df = pd.read_csv("output_data/cities.csv")

# Display sample data
city_data_df.head()
```

---

### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display

# Configure the map plot
map_plot = city_data_df.hvplot.points(
    'Lng', 
    'Lat', 
    geo=True, 
    tiles='OSM', 
    alpha=.6,
    size=30, 
    color='City', 
    frame_width=800, 
    frame_height=600
)

# Display the map
map_plot
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values
city_filtered = city_data_df[
    (city_data_df["Max Temp"] <= 29.444) &
    (city_data_df["Max Temp"] >= 10) &
    (city_data_df["Humidity"] >= 30) &
    (city_data_df["Humidity"] <= 60) &
    (city_data_df["Wind Speed"] <= 5.4) &
    (city_data_df["Cloudiness"] <= 60)
]

# Drop any rows with null values
city_filtered = city_filtered.dropna()

# Display sample data
city_filtered.head(10)
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = city_filtered.copy()
hotel_df = hotel_df[["City", "Country", "Lat", "Lng", "Humidity"]]

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df["Hotel Name"] = np.nan  # Use np.nan to initialize with NaN values
hotel_df["Hotel Name"] = hotel_df["Hotel Name"].astype(object)

# Display sample data
hotel_df.head()
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
radius = 10000
params = {
    "categories": "accommodation.hotel",
    "apiKey": geoapify_key, 
    "limit": 1
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    latitude = row['Lat']
    longitude = row['Lng']

    # Create a copy of the params dictionary for the current city
    current_params = params.copy()

    # Update the current_params dictionary for the current city
    current_params["filter"] = f"circle:{longitude},{latitude},{radius}"
    current_params["bias"] = f"proximity:{longitude},{latitude}"

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the current_params dictionary
    name_address = requests.get(base_url, params=current_params)

    # Convert the API response to JSON format
    name_address = name_address.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
%%capture --no-display

# Configure the map plot
hotel_plot = hotel_df.hvplot.points( 
    'Lng', 
    'Lat', 
    geo=True, 
    tiles='OSM', 
    alpha=.6, 
    size=30, 
    color='City', 
    frame_width=800, 
    frame_height=600, 
    hover_cols=['City', 'Country', 'Hotel Name'] )


# Display the map
hotel_plot
```
